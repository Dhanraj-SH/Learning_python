```python
#joining of the sets

#union() method
set0 = {"a", "b", "c"}
set1 = {1, 2, 3}
set2 = set0.union(set1)
print(set2)

#instead of union() we can also use "|" which is gonna give the same results
set3 = set1 | set2
print(set3)

#joining of multiple sets using union() method or "|"
set4 = set0.union(set1,set2,set3)
print(set4)
set5 = set0 | set1 | set2 | set3 | set4
print(set5)

#joining set with tuple and list
list0 = ("apple", "banana", "cherry")
tuple0 = ("mango", "kiwi", "strawberry")
set6 = set5.union(list0,tuple0)
print(set6)

#update() method
set7 = {4, 5, 6}
set7.update(set6)
print(set7)

#intersection() method 
# the method only returns the element which are present in the both the sets
set8 = {4, 5, 6}
set9 = set8.intersection(set7)
print(set9)

#and we can also use & operator instead of intersection() method
set10 = set9 & set8
print(set10)

#intesection_update() method
#this method also only keeps duplicates but it will change the original set instead of returning a new set
set11 = set6.copy()

print(set11)
set11.intersection_update(set8)
print(set11)

set12 = {"apple", 1,  "banana", 0, "cherry"}
set13 = {False, "google", 1, "apple", 2, True}

set14 = set12.intersection(set13)
print(set14)

#difference() method 
#difference method will return the item that are only present in the first set not present in the second set

set15 = set12.difference(set13)
print(set15)

# we can use "-" which will be returning the same output as the difference() method
set16 = set12 - set13
print(set16)

set17 = {"apple", "banana", "cherry"}
set18 = {"google", "microsoft", "apple"}

set17.difference_update(set18)
print(set17)

#symmetric_difference() 
#the symmetric difference will keep the elements which are not present in both the set

set19 = {"apple", "banana", "cherry"}
set20 = {"google", "microsoft", "apple"}

set21 = set19.symmetric_difference(set20)
print(set21)

#we can use "^" opreator instead of the symmetric_diffrence() method
set22 = set19 ^ set20

#symmetric_difference_update() method
#this method will also keep all the duplicates but it will change the original set

set19.symmetric_difference_update(set20)
print(set19)
```

Every recursive function must have 2 parts

1. A base case - condition that stops the recursion 
2. A recursive case - condition calling itselgf with a modified arguments


```python
#factorail problem using recursion

def factorail(n):
    #Base case
    if n == 0 or n == 1:
        return 1
    #recursive case
    else:
        return n * factorail(n-1)
    
print(factorail(5))

#fibonacci sequence problem using recursion 

def fibonaacci(n):
    if n<=1:
        return n
    else:
        return fibonaacci(n-1) + fibonaacci(n-2)
    
print(fibonaacci(7))

#recursion with lists

def sum_list(numbers):
    if len(numbers) == 0:
        return 0
    else:
        return numbers[0] + sum_list(numbers[1:])
    
numbers = [1,2,3,4]    
print(sum_list(numbers))

#maximum numbers in the list

def max_num(numbers):
    if len(numbers) == 1:
        return numbers[0]
    else:
        max_rest = max_num(numbers[1:])
        return numbers[0] if numbers[0] > max_rest else max_rest
    
numbers = [5,64,543,4,0]
print(max_num(numbers))

#minimum number in the list

def min_num(numbers):
    if len(numbers) == 1:
        return numbers[0]
    else:
        min_rest = min_num(numbers[1:])
        return numbers[0] if numbers[0] < min_rest else min_rest
    
numbers = [5,64,543,4,0]
print(min_num(numbers))
```

PYTHON GENERATORS

Generators 

Generators are functions that can pause and resume their execution.
When a generator function is called, it returns a genrator object, which is an iterator.
The code inside the function is not exectued yet, it is only complied. The function only executes when you iterate over the generator

```python
def generator():
    yield 1
    yield 2
    yield 3

for value in generator():
    print(value)


# the yield keyword is what it makes functions as a genrators.
# when yield is encountered, the function's state is saved, and the value is returned the next time the generator is called, it continues from where it left off

def count_upto(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for num in count_upto(5):
    print(num)

#generators saves memory
# generators are memory efficient because they generate values on the fly insted of storing everything in the memory
# for lage datasets, generators saves the memory 

def large_sequence(n):
    for i in range(n):
        yield i

#this doesn't create a million numbers in memory
gen = large_sequence(100000000)
print(next(gen))
print(next(gen))
print(next(gen))

#using next() with generators

#you can manually iterate through a generator using the next() function

def simple_gen():
    yield "Mary"
    yield "Mark"
    yield "Mattew"
    
gen = simple_gen()
print(next(gen))
print(next(gen))
print(next(gen))

#Generator expression 

list0 = [x * x for x in range(5)]
print(list0)

gen_exp = (x * x for x in range(5))
print(gen_exp)
print(builtins.list(gen_exp))

#using generator function with sum

total = sum(x * x for x in range(10))
print(total)
```

```python
#Fibonacci sequence generator

def fibonaacci():
    a, b = 0, 1
    while True:
        yield a 
        a, b = b, a + b

gen = fibonaacci()
for _ in range(15):
    print(next(gen))
```

```python
#Generator methods

#send() method - it allows u to send a value to the generator

def echo_gen():
    while True:
        received = yield
        print("Received:", received)

gen = echo_gen()
next(gen)
gen.send("Hello")
gen.send("World!")

#close() method - this method stops the generator

def my_gen():
  try:
    yield 1
    yield 2
    yield 3
  finally:
    print("Generator closed")

gen = my_gen()
print(next(gen))
gen.close()
```

Create an Iterator

To create an object/class as an iterator you have to implement the methods __iter__() and __next__() to your object.

As you will learn in the Python Classes/Objects chapter, all classes have a function called __init__(), which allows you to do some initializing when the object is being created.

The __iter__() method acts similar, you can do operations (initializing etc.), but must always return the iterator object itself.

The __next__() method also allows you to do operations, and must return the next item in the sequence.

```python
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    x = self.a
    self.a += 1
    return x

myclass = MyNumbers()
myiter = iter(myclass)

print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
```

StopIteration

The example above would continue forever if you had enough next() statements, or if it was used in a for loop.

To prevent the iteration from going on forever, we can use the StopIteration statement.

In the __next__() method, we can add a terminating condition to raise an error if the iteration is done a specified number of times:

```python
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    if self.a <= 20:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
  print(x)
```

```python
#built-in math functions
#the min() and max() functions can be used to find lowest and highest value in an iterable:

a = min(5, 10, 25)
b = max(5, 10, 25)

print(a)
print(b)

c = abs(-7.25) #the abs() function returns the absolute (positive) value of the specified number

print(c)

d = pow(4, 3) #The pow(x, y) function returns the value of x to the power of y

print(d)

#THE MATH MODULE

#python has also a built-in module called math, which extends the list of mathematical functions.
#To use it, you must import the math module:

#When u have imported the math module, you can start using methods and constants of the module
#The math.sqrt() method for example, returns the square root of a number

import math
print(math.sqrt(64))

print(math.ceil(1.4))# the math.ceil() method rounds a number upward to its nearest integer

print(math.floor(1.4))# the math.floor() method rounds a number downwards to its nearest integer 

print(math.pi)# the math.pi constant, returns the value of pi 
```

PYTHON JSON

JSON is a syntax for storing and exchanging data.
JSON is text, written with JavaScript object notation.

JSON in python

Python as a built-in package called json, which can be used to work with JSON data.

```python
#Prase json - convert from JSON to python
#if you have a json string, you can pause it by using the json.loads() method

import json

x = '{"name":"John", "age":30, "city":"New York"}' #some JSON:
y = json.loads(x) # prase x:

print(y["age"]) # the result is a Python dictionary:

#Convert from Python to JSON
#If you have a python object, you can convert it into a JSON string by using the json.dumps() method.

a = {
    "name" : "John" , 
    "age" : 30,
    "city" : "New York"
}# a Python object
b = json.dumps(x) #convert into JSON

print(b) #the result is a json string:

'''
You can convert Python objects of the following types, into JSON strings:

dict
list
tuple
string
int
float
True
False
None
'''

#When u convert from python to json, python objects are converted into the JSON (javascript) equivalent:
print(json.dumps({"name": "John", "age": 30}))#Object
print(json.dumps(["apple", "bananas"]))#Array
print(json.dumps(("apple", "bananas")))#Array
print(json.dumps("hello"))#String
print(json.dumps(42))#Number
print(json.dumps(31.76))#Number
print(json.dumps(True))#true
print(json.dumps(False))#false
print(json.dumps(None))#null

c = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

print(json.dumps(c))

#Format the result

#The example above prints a JSON string, but it is not very easy to read, with no indentations and line breaks
#The json.dumps() method has parameters to make it easier to read the result

json.dumps(c, indent=4) #use the indent parameter to define the number of indents

#you can also define the separators, default value is (", ", ": ")
# which means using a comma and a space to separate each object, and a colon and a space to seperate keys from values:

json.dumps(a, indent=4, separators=(". ", " = "))

#Order the results
#the json.dumps() method has parameters to order the keys in the result:

json.dumps(x, indent=4, sort_keys=True)#Use the sort_keys parameters to specify if the result should be sorted or not
```

PYTHON RegEx

A RegEx, or Regular Expression, is a sequence of characters that forms a search pattern.

RegEx can be used to check if a string contains the specified search pattern.


```python
#RegEx Module
#python has a built-in package called re, which can be used to work with Regular Expressions.

#RegEx in python 
#When you have imported the re module you can start with "The" and ends with "Spain"

import re#import th re module

txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt)
```

RegEx Functions - The re module offers a set of functions that allows us to search a string for a match:

Function	Description

findall-    Returns a list containing all matches

search-	    Returns a Match object if there is a match anywhere in the string

split-	    Returns a list where the string has been split at each match

sub-        Replaces one or many matches with a string


Metacharacters - metacharacters are characters with a special meaning

Character	Description	                                                                    Example	

[]	        A set of characters	                                                            "[a-m]"	

\	        Signals a special sequence (can also be used to escape special characters)	    "\d"	

.	        Any character (except newline character)	                                    "he..o"	

^	        Starts with	                                                                    "^hello"	

$	        Ends with	                                                                    "planet$"	

*	        Zero or more occurrences	                                                    "he.*o"	

+	        One or more occurrences	                                                        "he.+o"	

?	        Zero or one occurrences	                                                        "he.?o"	

{}	        Exactly the specified number of occurrences	                                    "he.{2}o"	

|	        Either or	                                                                    "falls|stays"	

()	        Capture and group


Flags - You can add flags to the pattern when using regular expressions.

Flag	        Shorthand	    Description	

re.ASCII	    re.A	        Returns only ASCII matches	

re.DEBUG		                Returns debug information	

re.DOTALL	    re.S	        Makes the . character match all characters (including newline character)	

re.IGNORECASE	re.I	        Case-insensitive matching	

re.MULTILINE	re.M	        Returns only matches at the beginning of each line	

re.NOFLAG		                Specifies that no flag is set for this pattern	

re.UNICODE	    re.U	        Returns Unicode matches. This is default from Python 3. For Python 2: use this flag to return only Unicode matches	

re.VERBOSE	    re.X	        Allows whitespaces and comments inside patterns. Makes the pattern more readable


There are still concepts of special sequences and sets

```python
import re

txt = "The rain in Spain"

#the findall() function returns a list containing all the matches
x = re.findall("ai", txt)
print(x)

#if no matches were found
xn = re.findall("USA", txt)
print(xn)


#the search function searches the string for a match, and returns a Match object if there is a match
#if there is more than one match, only the first occurrence of the match will be returned
y = re.search("\s", txt)
print("The first white-space character is located in position:",y.start())

#if no matches were found
yn = re.search("USA", txt)
print(yn) 



#the split function() returns a list where the string has been split at each match
z = re.split("\s", txt)
print(z) 

#u can control the number of occurrences by specifiying the maxsplit parameter
zn = re.split("\s", txt, 1)
print(zn)



#the sub() function replaces the matches with the text of your choice
a = re.sub("\s", "9", txt)
print(a)

#you can control the number of replacements by specifiying the count parameter
an = re.sub("\s", "9", txt, 2)
print(an)
```

```python
#MATCH OBJECT
#a match object is an object containing the information about the search and the result

import re

txt = "The rain in Spain"

'''
The match object has properties and methods used to retrive information about the search, and the result:

.span() returns a tuple containing the start-, and end postions of the match.
.string returns the string passed into the function
.group() returns the part of the string where there was a match

'''

x = re.search(r"\bS\w+", txt)#The regular expression looks for anyword that starts with the upper case 'S'
print(x.span())#prints the postion(strat and end position value) of the first occurrence match



y = re.search(r"\bS\w+", txt)
print(y.string)#prints the string passed into the function



z = re.search(r"\bS\w+", txt)
print(z.group())#prints the part of the string where from where the match starts 
```

PYTHON TRY EXCEPT

The try block lets you test a block of code for errors.

The except block lets you handle the error.

The else block lets you execute code when there is no error.

The finally block lets you execute code regardless the result of the try and except block


```python
'''

Exception handling
When an error occures, or exception as we call it, Python will normally stop and generate an error msg
These exceptions can be handeled using try statment

'''
try:
    print(m)
except:
    print("An exception occured")

#Since the try block will generate error bcz of the undefined variable m it expect block will executed

#Many exception

try:
    print(m)
except NameError:
    print("Variable x is not defined")
except:
    print("Something else went wrong")



#we can use else keyword to define a block of code to be executed if no errors were raisesd
try:
  print("Hello")
except:
  print("Something went wrong")
else:
  print("Nothing went wrong")



# The finally block, if specified, will be executed regardless if the try block rasies an error or not
try:
  print(m)
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")


try:
  f = open("demofile.txt")
  try:
    f.write("Lorum Ipsum")
  except:
    print("Something went wrong when writing to the file")
  finally:
    f.close()
except:
  print("Something went wrong when opening the file")


```

```python

'''
Raise an exception
As a python developer you can choose to throw an exception if a condition occurs.
to throw (or raise) an exception, use the raise keyword
'''

x = -1

if x < 0:
   raise Exception("Sorry, no numbers below zero") #the raise keyword raise an exception
```

PYTHON STRING FORMATTING

F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.

Before Python 3.6 we had to use the format() method.

```python
'''
F-String
it allows you to format selected part of a string
To specify a string as a f-string simply put an f in front of the string literal, like this
'''
txt = f"The price is 49 dollars"
print(txt)



'''
Placholders and Modifiers
To format values in an f-string, add placeholders {}, a placeholder can contain variables, operations, functions and modifiers to format the value.
'''
price = 59
txt = f"The price is {price} dollars"
print(txt)



'''
A placeholder can also include a modifier to format the value.
A modifier is included by adding a colon : followed by a legal formatting type, like .2f which means fixed point number with 2 decimals:
'''
txt = f"The price is {price:.2f} dollars"
print(txt)

txt = f"The price is {95:.2f} dollars"
print(txt)
```
