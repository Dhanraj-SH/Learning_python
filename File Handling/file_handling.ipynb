# PYTHON FILE HANDLING

File handling is a important part of any web applications

Python has serveral functions for creating, reading, writing and deleting files.

#### FILE HANDLING

The Key function for working with files in python is the `open()` function.

The `open()` function takes 2 parameters: filename and mode.

There are four different methods(modes) for opening a file:

| Mode | Name   | Description |
|------|--------|------------|
| "r"  | Read   | Default value. Opens a file for reading, error if the file does not exist |
| "a"  | Append | Opens a file for appending, creates the file if it does not exist         |
| "w"  | Write  | Opens a file for writing, creates the file if it does not exist           |
| "x"  | Create | Creates the specified file, returns an error if the file exists           |
|      |        |                                                                           |
|      |        | In addition you can specify if the file sholud be handled as binary or txt|
|      |        |                                                                           |
| "t"  | Text   | Default value. Text mode                                                  |
| "b"  | Binary | Binary mode(e.g. images)                                                  |



#### SYNTAX 
 
f = open("demofile.txt")

The code above is the same as:

f = open("demofile.txt", "rt")

Because "r" for read, and "t" for text are the default values, you do not need to specify them

#### PYTHON FILE OPEN

```python
# Open a file on the server
f = open("demofile.txt") # when the file is located in the same folder we can use the name of the file directly else we need to specify the file path 
print(f.read())



#Using the with statment
with open("demofile.txt") as f:
    print(f.read()) # when we use 'with' statment we need not to wry abt the closing the file 'with' statement takes care of that



'''
Close Files

It is a good practice to always close the file when you are done with it.
If we are not using the 'with' statment, you must write a close statment takes care of the file

You should always close your files. 
In some cases, due to buffering, changes made to a file may not show until you close the file.
'''
f = open("demofile.txt")
print(f.read())
f.close()


#Read only Parts of file - By default the read() method returns the whole text, but you can also specify how many characters you want to return:
with open("demofile.txt") as f:
  print(f.read(5))


#Read Lines - We can return one line by using the readline() method
with open("demofile.txt") as f:
   print(f.readline())


#by looping through the lines of the file, you can read the whole file line by line:
with open("demofile.txt") as f:
  for x in f:
    print(x)
```
