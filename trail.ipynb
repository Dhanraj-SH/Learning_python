```python
#joining of the sets

#union() method
set0 = {"a", "b", "c"}
set1 = {1, 2, 3}
set2 = set0.union(set1)
print(set2)

#instead of union() we can also use "|" which is gonna give the same results
set3 = set1 | set2
print(set3)

#joining of multiple sets using union() method or "|"
set4 = set0.union(set1,set2,set3)
print(set4)
set5 = set0 | set1 | set2 | set3 | set4
print(set5)

#joining set with tuple and list
list0 = ("apple", "banana", "cherry")
tuple0 = ("mango", "kiwi", "strawberry")
set6 = set5.union(list0,tuple0)
print(set6)

#update() method
set7 = {4, 5, 6}
set7.update(set6)
print(set7)

#intersection() method 
# the method only returns the element which are present in the both the sets
set8 = {4, 5, 6}
set9 = set8.intersection(set7)
print(set9)

#and we can also use & operator instead of intersection() method
set10 = set9 & set8
print(set10)

#intesection_update() method
#this method also only keeps duplicates but it will change the original set instead of returning a new set
set11 = set6.copy()

print(set11)
set11.intersection_update(set8)
print(set11)

set12 = {"apple", 1,  "banana", 0, "cherry"}
set13 = {False, "google", 1, "apple", 2, True}

set14 = set12.intersection(set13)
print(set14)

#difference() method 
#difference method will return the item that are only present in the first set not present in the second set

set15 = set12.difference(set13)
print(set15)

# we can use "-" which will be returning the same output as the difference() method
set16 = set12 - set13
print(set16)

set17 = {"apple", "banana", "cherry"}
set18 = {"google", "microsoft", "apple"}

set17.difference_update(set18)
print(set17)

#symmetric_difference() 
#the symmetric difference will keep the elements which are not present in both the set

set19 = {"apple", "banana", "cherry"}
set20 = {"google", "microsoft", "apple"}

set21 = set19.symmetric_difference(set20)
print(set21)

#we can use "^" opreator instead of the symmetric_diffrence() method
set22 = set19 ^ set20

#symmetric_difference_update() method
#this method will also keep all the duplicates but it will change the original set

set19.symmetric_difference_update(set20)
print(set19)
```

```python
#Nested dictonary
name_list = {
    "child0" : {
        "name" : "Ananth",
        "year" : 2003
    },
    "child1" : {
        "name" : "Dhanraj",
        "year" : 2004
    },
    "child2" : {
        "name" : "Naveen",
        "year" : 2005
    }
}
print(name_list["child2"]["name"])


#another way to create a nested dictonary
child4 = {
    "name" : "Ananth",
    "year" : 2003
}
child5 = {
    "name" : "Dhanraj",
    "year" : 2004
}
child6 = {
    "name" : "Naveen",
    "year" : 2005
}
name_list1 = {
    "child4" : child4,
    "child5" : child5,
    "child6" : child6
}
print(name_list1["child4"]["name"])

#looping through the nested dictonary
for x,obj in name_list1.items():
    print(x)

    for y in obj:
        print(y + ':', obj[y])
```

arbitrary arguments - *args

if u don't know how many arguments will be passed into your function, add a *before the parameter this way funtion will receive a tuple of arguments and can access the items accordingly


```python
def fun0(*kids):
    print("The youngest child is " + kids[2])

fun0("Emil", "Tobias", "Linus")

#the *args parameter allows a function to accept any number of positional arguments 
#Inside the function, args becomes a tuple containing all the passed arguments

def fun1(*kids):
    print("Type:", type(kids))
    print("First argument:", kids[0])
    print("Second argument:", kids[1])
    print("All arguments",kids)

fun1("Emil", "Tobias", "Linus")

#Using *args with regular arguments
def fun2(greeting, *names):
    for name in names:
        print(greeting, name)

fun2("Hello", "Emil", "Tobias", "Linus")

#Practical example for *args

def fun3(*nums):
    total = 0
    for num in nums:
        total += num 
    return total 

print(fun3(1,2,3))
print(fun3(10,20,30,40))
print(fun3(5))

#Lets using args for finding the maximum number

def fun4(*nums):
    if len(nums) == 1 | len(nums) == 0:
        return nums;
    max = 0
    for num in nums:
        if max < num:
            max = num
    return max

print(fun4(3,45,5,563,5,3,53))    
```
